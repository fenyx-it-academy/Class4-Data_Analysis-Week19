```python
df_titanic = pd.read_csv('/kaggle/input/titanic/train.csv')
df_titanic.head(1)
```


Task: Create a multi-indexed dataframe

Indexes:

level_0 = Sex level_1 = Embarked Columns:

level_0 = Sex level_1 = Survived Values: Random

```python
index = [df_titanic.Sex.unique(), df_titanic.Embarked.unique()]
index = pd.MultiIndex.from_product(index, names=["Sex", "Embarked"])
column = [df_titanic.Pclass.unique(), df_titanic.Survived.unique()]
column = pd.MultiIndex.from_product(column, names=["Pclass", "Survived"])
df1 = pd.DataFrame(np.random.randn(8,6), index=index, columns=column)
df1
```

```python
# QUESTION 2  

# Task: Create pivot table by using pivot
# Index: Name
# Column: Ticket
# Values: Age

df2= df_titanic.iloc[:10].pivot(index='Name',columns='Ticket',values='Age')
df2
```

```python
# QUESTION 3
# Task: Manipulate the table you obtained in the 1st task with the stack and unstack methods

df3 = df1.stack([0,1]).unstack([0,3])
df3
```

```python
# QUESTION 4

# Task: Create the table below using melt method
# Id_vars: 'Name' and 'Embarked'
# Value_vars: 'Sex' and 'Age'

pd.melt(df_titanic.iloc[:10],id_vars= ['Name', 'Embarked'],value_vars= ['Sex', 'Age'],var_name='variable', value_name='value')
```

```python
# QUESTION 5

# Task: Create the following pivot tables using the pivot_table method
# Index:
# •	A) 'Embarked'
# •	B) 'Embarked', 'Sex', 'Pclass'
# •	C) 'Embarked', 'Sex', 'Pclass'
# Values: 'Fare' and 'Age'
# aggfunc:
# •	A) 'Fare': mean, 'Age': median
# •	B) 'Fare': mean, 'Age': median
# •	C) 'Fare': mean and np.sum, 'Age': np.median and np.std


pd.pivot_table(df_titanic, values=['Age','Fare'], index="Embarked",aggfunc={'Age' : np.median,'Fare':np.mean})
```

```python
pd.pivot_table(df_titanic, values=['Age','Fare'], index=['Embarked', 'Sex', 'Pclass'],aggfunc={'Age' : np.median,'Fare':np.mean})
```

```python
pd.pivot_table(df_titanic, values=['Age','Fare'], index=['Embarked', 'Sex', 'Pclass'],aggfunc={'Age' : [np.median,np.std],'Fare':[np.mean,np.sum]})
```
