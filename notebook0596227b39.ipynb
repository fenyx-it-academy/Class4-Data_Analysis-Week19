### 5. Pivot Table:

__Data:__ https://www.kaggle.com/c/titanic

__Task:__ Create the following pivot tables using the pivot_table method

__Index:__ 
* A) 'Embarked'
* B) 'Embarked', 'Sex', 'Pclass'
* C) 'Embarked', 'Sex', 'Pclass'

__Values:__ 'Fare' and 'Age'

__aggfunc:__
* A) 'Fare': mean, 'Age': median
* B) 'Fare': mean, 'Age': median
* C) 'Fare': mean and np.sum, 'Age': np.median and np.std


```python
pivot_A = df_tianic.pivot_table(index="Embarked", values=["Fare","Age"],  aggfunc={"Fare":np.mean,"Age": np.median})
pivot_A
```

```python
pivot_B = df_tianic.pivot_table(index=["Embarked", "Sex", "Pclass"], values=["Fare","Age"],  aggfunc={"Fare": np.mean, "Age":np.median})
pivot_B
```

```python
pivot_B = df_tianic.pivot_table(index=["Embarked", "Sex", "Pclass"], values=["Fare","Age"],  aggfunc={"Fare": [np.mean, np.sum], "Age":[np.median, np.std]})
pivot_B
```
