```python
df = pd.read_csv('/kaggle/input/titanic/train.csv')
df.head(1)
```

```python
# 1. Multi index & column:
# Data: https://www.kaggle.com/c/titanic
# Task: Create a multi-indexed dataframe
# Indexes:
# level_0 = Sex
# level_1 = Embarked
# Columns:
# level_0 = Sex
# level_1 = Survived
# Values: Random

a=df.copy()
```

```python
#  Question-1
ind1=list(set (list(df.Sex)))
ind2=list(set (list(df.Embarked)))
col1=list(set (list(df.Pclass)))
col2=list(set (list(df.Survived)))
index = pd.MultiIndex.from_product([ind1,ind2], names=["Sex", "Embarked"])
columns = pd.MultiIndex.from_product([col1,col2], names=["Pclass", "survived"])
b = pd.DataFrame(np.random.randn(8,6), columns=columns, index=index)
b
```

```python
# 2.Question
a.iloc[:10].pivot(index='Name', columns='Ticket', values='Age')
```

```python
# Question-3
b.stack(0).stack().unstack(0).unstack()
```

```python
# Question-4
pd.melt(df.iloc[:10],id_vars= ['Name', 'Embarked'],value_vars= ['Sex', 'Age'],var_name='variable', value_name='value')
```

```python
# Question-5
pd.pivot_table(a, values=['Fare','Age'], index="Embarked",aggfunc={'Age':np.median,'Fare':np.mean})
```

```python
pd.pivot_table(a, values=['Fare','Age'], index=["Embarked", 'Sex', 'Pclass'],aggfunc={'Age':np.median,'Fare':np.mean})
```

```python
pd.pivot_table(a, values=['Fare','Age'], index=["Embarked", 'Sex', 'Pclass'],aggfunc={'Fare': [np.mean, np.sum], 'Age': [np.median, np.std]}) 
```
