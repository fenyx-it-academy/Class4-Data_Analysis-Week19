```python
#1
ind = [df.Sex.unique(), df.Embarked.unique()]
index = pd.MultiIndex.from_product(ind, names=["Sex", "Embarked"])
col = [df.Pclass.unique(), df.Survived.unique()]
column = pd.MultiIndex.from_product(col, names=["Pclass", "Survived"])
df1 = pd.DataFrame(np.random.randint(9,size=(8,6)), index=index, columns=column)
df1
```

```python
#2
df2 = df.iloc[:10].pivot(index='Name',columns='Ticket',values='Age')
df2
```

```python
#3
df3 = df1.stack([0,1]).unstack([0,3])
df3
```

```python
#4
df4 = df.melt(id_vars=['Name','Embarked'], value_vars =['Sex', 'Age'], var_name='variable', value_name='value').iloc[:10]
df4
```

```python
#5
pd.pivot_table(df, values=['Fare','Age'], index="Embarked",aggfunc={'Fare':[np.mean,np.sum], 'Age' : [np.median,np.std]})
```
