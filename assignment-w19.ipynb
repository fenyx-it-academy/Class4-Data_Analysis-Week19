# 5. Pivot Table:
# Task: Create the following pivot tables using the pivot_table method

- Index:
- A) 'Embarked'
- B) 'Embarked', 'Sex', 'Pclass'
- C) 'Embarked', 'Sex', 'Pclass'
- Values: 'Fare' and 'Age'
- aggfunc:
- A) 'Fare': mean, 'Age': median
- B) 'Fare': mean, 'Age': median
- C) 'Fare': mean and np.sum, 'Age': np.median and np.std

```python
task5A = pd.pivot_table(df,values=['Age','Fare'], index='Embarked',aggfunc={'Age' : np.median,'Fare':np.mean})
task5A
```

```python
task5B = pd.pivot_table(df,values=['Age','Fare'],index=['Embarked','Sex','Pclass'],aggfunc={'Fare':np.mean,'Age':np.median})
task5B
```

```python
task5C = pd.pivot_table(df, values=['Age','Fare'], index=['Embarked', 'Sex', 'Pclass'],aggfunc={'Age' : [np.median,np.std],'Fare':[np.mean,np.sum]})
task5C
```
