# 5.Sorunun Cevabi (Create the following pivot tables using the pivot_table method)

```python
A=pd.pivot_table(df_titanic, values=['Age','Fare'], index="Embarked",aggfunc={'Age' : np.median,'Fare':np.mean})

A

```

```python
B=pd.pivot_table(df_titanic, values=['Age','Fare'], index=['Embarked', 'Sex', 'Pclass'],aggfunc={'Age' : np.median,'Fare':np.mean})
B
```

```python
C=pd.pivot_table(df_titanic, values=['Age','Fare'], index=['Embarked', 'Sex', 'Pclass'],aggfunc={'Age' : [np.median,np.std],'Fare':[np.mean,np.sum]})
C
```
