
Pivot Table: Data: https://www.kaggle.com/c/titanic
Task: Create the following pivot tables using the pivot_table method

Index:

A) 'Embarked'

B) 'Embarked', 'Sex', 'Pclass'

C) 'Embarked', 'Sex', 'Pclass'

Values: 'Fare' and 'Age'

aggfunc:

A) 'Fare': mean, 'Age': median

B) 'Fare': mean, 'Age': median

C) 'Fare': mean and np.sum, 'Age': np.median and np.std

```python
# A) 'Embarked'
pd.pivot_table(tdf, values=["Fare","Age"], index=["Embarked"], aggfunc={'Fare':[np.mean], 'Age':[np.median]})
```

```python
pd.pivot_table(tdf, values=["Fare","Age"], index=["Embarked", 'Sex', 'Pclass'], aggfunc={'Fare':[np.mean,np.sum], 'Age':[np.median,np.std]})
```

```python
# ) B) 'Embarked', 'Sex', 'Pclass', B) 'Fare': mean, 'Age': median

pd.pivot_table(tdf, values=["Fare","Age"], index=["Embarked", 'Sex', 'Pclass'], aggfunc={'Fare':[np.mean], 'Age':[np.median]})
```

```python
# C)  'Embarked', 'Sex', 'Pclass' C)'Fare': mean and np.sum, 'Age': np.median and np.std
```

```python
pd.pivot_table(tdf, values=["Fare","Age"], index=["Embarked", 'Sex', 'Pclass'], aggfunc={'Fare':[np.mean,np.sum], 'Age':[np.median,np.std]})
```
