# 5. Pivot Table:
Data: https://www.kaggle.com/c/titanic

Task: Create the following pivot tables using the pivot_table method

Index:

* A) 'Embarked'
* B) 'Embarked', 'Sex', 'Pclass'
* C) 'Embarked', 'Sex', 'Pclass'

Values: 'Fare' and 'Age'

aggfunc:

* A) 'Fare': mean, 'Age': median
* B) 'Fare': mean, 'Age': median
* C) 'Fare': mean and np.sum, 'Age': np.median and np.std

```python
piv_A = pd.pivot_table(dfc, values = ['Age','Fare'], index = "Embarked",aggfunc = {'Age' : np.median,'Fare':np.mean})
piv_A
```

```python
piv_B = pd.pivot_table(dfc, values = ['Age','Fare'], index = ['Embarked', 'Sex', 'Pclass'],aggfunc = {'Age' : np.median,'Fare':np.mean})
piv_B
```

```python
piv_c = pd.pivot_table(df, values = ['Age','Fare'], index = ['Embarked', 'Sex', 'Pclass'],aggfunc = {'Age' : [np.median,np.std],'Fare':[np.mean,np.sum]})
piv_c
```
